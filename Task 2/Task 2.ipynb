```python
import keras
from keras.datasets import mnist
(train_X, train_y), (test_X, test_y) = mnist.load_data()
```

```python
from matplotlib import pyplot
pyplot.figure(2, figsize=(4, 4))
pyplot.imshow(train_X[5], cmap=pyplot.get_cmap('gray'))
pyplot.show()
```

```python
import sklearn.svm as svm
```

```python
train_X = train_X.reshape((60000,28*28))
test_X = test_X.reshape((10000,28*28))  
train_X = train_X.astype('float32')/255
test_X = test_X.astype('float32')/255
```

```python
test_X.shape
```

```python
test_y.shape
```

```python
svc = svm.LinearSVC(C = 1, loss = "hinge", max_iter = 10000)
```

```python
svc.fit(train_X,train_y)
```

```python
svc.predict(test_X[3].reshape(1,-1))
```

```python
pyplot.figure(2, figsize=(4, 4))
pyplot.imshow(test_X[3].reshape(28,28), cmap=pyplot.get_cmap('gray'))
pyplot.show()
```

```python
svc.predict(test_X[345].reshape(1,-1))
```

```python
pyplot.figure(2, figsize=(4, 4))
pyplot.imshow(test_X[345].reshape(28,28), cmap=pyplot.get_cmap('gray'))
pyplot.show()
```

```python
from  sklearn.metrics import plot_confusion_matrix 
plot_confusion_matrix(svc, test_X, test_y)
```

```python
svc.score(test_X, test_y)
```
