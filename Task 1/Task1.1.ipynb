Аттрактор Лоренца

$ \begin{cases}
   \dot{x} = \sigma(y - x)       \\
   \dot{y} = x(r - z) - y \\
   \dot{z} = xy - bz
  \end{cases}
$

```python
import numpy as np
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

import bokeh.plotting as bp
from bokeh.io import output_notebook

def F1(t,s):
    sigma=10; r=28; b=2.667
    x,y,z = s
    return sigma*(y-x), x*(r-z)-y,x*y-b*z

sol = solve_ivp(F1,(-100,15),(0.,1.,1.05))
xs,ys,zs = sol.y
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(xs, ys, zs, lw=0.3)
```

```python
fig = plt.figure()

plt.subplot(2, 2, 1)
plt.plot(sol.t, xs,lw = 0.5)

plt.subplot(2, 2, 2)
plt.plot(sol.t, ys,lw = 0.5)

plt.subplot(2, 2, 3)
plt.plot(sol.t, zs,lw = 0.5)

plt.subplot(2, 2, 4)
plt.plot(xs,zs,lw = 0.5)

plt.show()
```

```python
def F(s,t):
    sigma=10; r=28; b=2.667
    dxdt = sigma*(s[1] - s[0])
    dydt = s[0]*(r - s[2]) - s[1]
    dzdt = s[0]*s[1] - b*s[2]
    return [dxdt,dydt,dzdt]

s0 = np.array([0.,1.,1.05])
t = np.linspace(-100,15,10000)
s = odeint(F,s0,t)
```

```python
xs = s[:,0]
ys = s[:,1]
zs = s[:,2]
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(xs, ys, zs, lw=0.5)
```

```python
ax = plt.plot(t,ys,lw = 0.5)
```

```python
ax = plt.plot(t,zs,lw = 0.5)
```
